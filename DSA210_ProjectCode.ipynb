```python
import pandas as pd

# Ekim verisi (1000 satır)
transport_oct = pd.read_csv(
    '/content/drive/MyDrive/DSA210_Project/hourly_transportation_202410.csv',
    usecols=['transition_date', 'transition_hour', 'number_of_passenger', 'line_name'],
    nrows=1000
)

traffic_oct = pd.read_csv(
    '/content/drive/MyDrive/DSA210_Project/traffic_density_202410.csv',
    usecols=['DATE_TIME', 'AVERAGE_SPEED'],
    nrows=1000
)

# Kasım verisi (1000 satır)
transport_nov = pd.read_csv(
    '/content/drive/MyDrive/DSA210_Project/hourly_transportation_202411.csv',
    usecols=['transition_date', 'transition_hour', 'number_of_passenger', 'line_name'],
    nrows=1000
)

traffic_nov = pd.read_csv(
    '/content/drive/MyDrive/DSA210_Project/traffic_density_202411.csv',
    usecols=['DATE_TIME', 'AVERAGE_SPEED'],
    nrows=1000
)

# Aralık verisi (1000 satır)
transport_dec = pd.read_csv(
    '/content/drive/MyDrive/DSA210_Project/hourly_transportation_202512.csv',  # dosya adı 2025 ama içeriği Aralık diye konuşmuştuk
    usecols=['transition_date', 'transition_hour', 'number_of_passenger', 'line_name'],
    nrows=1000
)

traffic_dec = pd.read_csv(
    '/content/drive/MyDrive/DSA210_Project/traffic_density_202412.csv',
    usecols=['DATE_TIME', 'AVERAGE_SPEED'],
    nrows=1000
)

print("✅ Her aydan 1000 satır veri başarıyla yüklendi!")
```

```python
# Toplu taşıma verilerini birleştir
transport = pd.concat([transport_oct, transport_nov, transport_dec], ignore_index=True)

# Trafik yoğunluk verilerini birleştir
traffic = pd.concat([traffic_oct, traffic_nov, traffic_dec], ignore_index=True)

print("✅ Üç ayın verileri birleştirildi!")
```

```python
# Public Transport DateTime ayarı
transport['date_time'] = pd.to_datetime(transport['transition_date'] + ' ' + transport['transition_hour'].astype(str) + ':00:00')

# Traffic Density DateTime ayarı
traffic = traffic.rename(columns={'DATE_TIME': 'date_time'})
traffic['date_time'] = pd.to_datetime(traffic['date_time'])

# Merge
data = pd.merge(transport, traffic, on='date_time')

print("✅ Veri başarıyla birleştirildi:")
display(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x='AVERAGE_SPEED', y='number_of_passenger', data=data)
plt.title('Traffic Average Speed vs Public Transport Usage')
plt.xlabel('Average Speed (km/h)')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python
# Verileri doğru tipe çevir (float veya int)
data['AVERAGE_SPEED'] = pd.to_numeric(data['AVERAGE_SPEED'], errors='coerce')
data['number_of_passenger'] = pd.to_numeric(data['number_of_passenger'], errors='coerce')

# Sonra korelasyonu tekrar hesapla
from scipy.stats import pearsonr

# Pearson correlation
corr_coef, p_value = pearsonr(data['AVERAGE_SPEED'], data['number_of_passenger'])

print(f"Pearson Correlation Coefficient: {corr_coef:.4f}")
print(f"P-Value: {p_value:.4f}")

# Hipotez Yorumu
alpha = 0.05
if p_value < alpha:
    print("✅ Sonuç: Null Hypothesis reddedildi. Trafik ortalama hızı ile toplu taşıma kullanımı arasında anlamlı bir ilişki var.")
    if corr_coef > 0:
        print("➡️ İlişki pozitiftir.")
    else:
        print("➡️ İlişki negatiftir.")
else:
    print("❌ Sonuç: Null Hypothesis reddedilemedi. Anlamlı bir ilişki bulunamadı.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sadece ihtiyacımız olan iki kolonu al
corr = data[['AVERAGE_SPEED', 'number_of_passenger']].corr()

# Heatmap çiz
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Average Speed vs Number of Passengers')
plt.show()
```

"trafik yoğunluğu" ölçümü doğrudan ortalama hız üzerinden yapılmış.
(İBB verisi bu şekilde: Trafik yoğunluğu değil, Average Speed verisi veriliyor.)

Mantık şu:

Ortalama hız azaldıkça trafik yoğunlaşır.
Ortalama hız arttıkça trafik rahatlar.
Özetle:

Düşük Average Speed → Yüksek Trafik Yoğunluğu demek,
Yüksek Average Speed → Düşük Trafik Yoğunluğu demek.
